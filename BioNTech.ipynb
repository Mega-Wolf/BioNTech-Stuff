```python
rna = open("BioNTech-mRNA.txt", "r")
aminosFile = open("Aminos.txt", "r")

codons = []
aminos = []
codonAminoLink = {}

skipped = False
for line in rna.readlines():   
    if not skipped:
        print(line)
        skipped = True
        continue
    divided = line.split(",")
    index = divided[0].strip()
    original = divided[1].strip()
    modified = divided[2].strip()
    codons.append((original, modified))
    
skipped = False
for line in aminosFile.readlines():
    if not skipped:
        print(line)
        skipped = True
        continue
        
    info = line.split(":")[0]
    restInfo = info.split(";")[1].split(",")
    
    typeName = info.split(";")[0]
    longName = restInfo[0]
    shortName = restInfo[1]
    letter = restInfo[2]
    
    possibilites = line.split(":")[1].strip().split(",")
    
    aminoLink = [longName, shortName, letter]
    aminos.append(aminoLink)
    
    for pos in possibilites:
        pos = pos.replace("U", "T")
        codonAminoLink[pos] = aminoLink

print(f"Amount codons: {len(codons)}")
uniqueCodonPairs = list(set(codons))
print(f"Amount different codon transfomrations: {len(uniqueCodonPairs)}")
print()

transforms = {}
specialTransforms = []

for amino in aminos:
    transforms[amino[2]] = []

for pair in uniqueCodonPairs:
    codonFrom = pair[0]
    codonTo = pair[1]
    aminoFrom = codonAminoLink[codonFrom][2]
    aminoTo = codonAminoLink[codonTo][2]
    amount = codons.count(pair)
    
    obj = [codonFrom, codonTo, aminoFrom, aminoTo, amount]
    
    if (aminoFrom == aminoTo):
        transforms[aminoFrom].append(obj)
    else:
        specialTransforms.append(obj)
    
    #print(f"{codonFrom}({aminoFrom}) -> {codonTo}({aminoTo}) : {amount}")


for amino in aminos:
    print(f"Transformations: {amino[0]} -> {amino[0]}")
    strings = []
    for data in transforms[amino[2]]:
        codonFrom = data[0]
        codonTo = data[1]
        aminoFrom = data[2]
        aminoTo = data[3]
        amount = data[4]
        
        if codonFrom == codonTo:
            strings.append(f"({codonFrom}({aminoFrom}) -> {codonTo}({aminoTo}) : {amount})")
        else:
            strings.append(f"{codonFrom}({aminoFrom}) -> {codonTo}({aminoTo}) : {amount}")

    strings.sort()
    for s in strings:
        print(s)
    print()
    
print(f"Special Transformations:")
for data in specialTransforms:
    codonFrom = data[0]
    codonTo = data[1]
    aminoFrom = data[2]
    aminoTo = data[3]
    amount = data[4]

    print(f"{codonFrom}({aminoFrom}) -> {codonTo}({aminoTo}) : {amount}")
print()
```
